```python
#Imports
import os
import pandas as pd
import numpy as np
```

```python
#Read CSV
ufo_data = pd.read_csv('../UFO_data.csv')
```

```python
#Remove American Territories - US States Only
ufo_data = ufo_data.drop(ufo_data[ufo_data['State'] == "American Samoa"].index)
```

```python
#Summary of UFO Sightings by State
print("Count of UFO Sightings by State:")
states = ufo_data.groupby(['State']).size().to_frame(name = 'Sightings').reset_index()
states
```

```python
# write to csv file
states.to_csv('sightings_state.csv')
```
