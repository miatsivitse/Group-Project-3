```python
#Imports
import os
import pandas as pd
import numpy as np
from datetime import datetime, date
```

```python
#Read CSV
ufo_data = pd.read_csv('../UFO_data.csv')
```

```python
#Remove American Territories - US States Only
ufo_data = ufo_data.drop(ufo_data[ufo_data['State'] == "American Samoa"].index)
```

```python
#Switch to Datetime and Normalize
ufo_data['Date'] = pd.to_datetime(ufo_data['Date']).dt.normalize()
ufo_data
```

```python
ufo_data.dtypes
```

```python
#Resample Months
ufo_months = ufo_data.set_index('Date').resample('M').count()
ufo_months = ufo_months.reset_index()
ufo_months
```

```python
#Summary of UFO Sightings by Month
print("Count of UFO Sightings by Month:")
ufo_monthly = ufo_months.groupby('Date').sum()['State'].to_frame(name = 'Sightings').reset_index()
ufo_monthly
```

```python
# write to csv file
ufo_monthly.to_csv('sightings_months.csv')
```
