```python
import pandas as pd 
```

```python
Dec2022 = pd.read_csv('Dec22_Data.csv')
Nov2022 = pd.read_csv('Nov22_Data.csv')
Oct2022 = pd.read_csv('Oct22_Data.csv')
Sep2022 = pd.read_csv('Sep22_Data.csv')
Aug2022 = pd.read_csv('Aug22_Data.csv')
Jul2022 = pd.read_csv('Jul22_Data.csv')

# combine all the datas
summary = pd.concat([Jul2022,Aug2022,Sep2022,Oct2022,Nov2022,Dec2022], axis=0)
```

```python
# rename and drop the first column
summary.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
summary.drop(["a"], axis=1, inplace=True)
```

```python
summary.head(5)
```

```python
# choose only USA
USummary = summary.loc[summary['Country'] == 'USA']
```

```python
# reset the index
USummary.reset_index(drop=True,inplace=True)
USummary
```

```python
# create new column ReportNum as per the index
USummary['ReportNum'] = (USummary.index)+1
```

```python
USummary.head(5)
```

```python
# write to csv file
USummary.to_csv('summary.csv')
```
